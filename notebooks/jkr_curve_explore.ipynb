# JKR pull-off work — parameter exploration

Quick visualization of how `W_adh^JKR(R, γ, K)` varies across grain radius, surface energy, and reduced modulus — the three free knobs of Operator A.

Useful for choosing config defaults in `data/configs/*.yaml` and for the `dry vs humid sand` demo (Step 5+).

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))

import numpy as np
import matplotlib.pyplot as plt
from src.coupling.jkr import jkr_pulloff_work_array, gamma_humid_np
```

## W_adh vs grain radius

Theory: W_adh ∝ R^(7/3) (loglog slope ≈ 2.33).

```python
R = np.geomspace(1e-4, 1e-2, 50)
for gamma_val, label in [(0.02, 'γ=0.02'), (0.05, 'γ=0.05'), (0.10, 'γ=0.10')]:
    w = jkr_pulloff_work_array(R, np.full_like(R, gamma_val), np.full_like(R, 1e6))
    plt.loglog(R, w, label=label)
plt.xlabel('grain radius R (m)')
plt.ylabel('W_adh (J)')
plt.title('JKR pull-off work vs grain radius')
plt.legend()
plt.grid(True, which='both')
plt.show()
```

## γ_humid(h, σ/σ_max)

Capillary boost should be largest at low saturation, vanish at full saturation. h=0 → γ_0.

```python
g0, beta, sigma_max = 0.05, 0.10, 0.05
h_vals = [0.0, 0.25, 0.5, 0.75, 1.0]
sig_frac = np.linspace(0.0, 1.0, 50)
for h in h_vals:
    gammas = [gamma_humid_np(g0, beta, h, sf * sigma_max, sigma_max) for sf in sig_frac]
    plt.plot(sig_frac, gammas, label=f'h={h}')
plt.xlabel('σ / σ_max')
plt.ylabel('γ (J/m²)')
plt.title('humidity-dependent γ vs saturation')
plt.legend()
plt.grid(True)
plt.show()
```
